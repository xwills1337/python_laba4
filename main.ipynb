```python
import pandas as pd
import datetime

df = pd.read_csv('data.csv', sep=';', encoding='cp1251')
df
```

```python
df.rename(columns={"Дата": "data", "День_температура": "day_temp", "День_давление": "day_press", "День_ветер": "day_wind", 
                  "Вечер_температура": "even_temp", "Вечер_давление": "even_press", "Вечер_ветер": "even_wind"}, inplace=True)
df
```

```python
df.loc[(df.day_temp == '-'), 'day_temp'] = 0
df.loc[(df.day_press == '-'), 'day_press'] = 0
df.loc[(df.day_wind == '-'), 'day_wind'] = 0
df.loc[(df.even_temp == '-'), 'even_temp'] = 0
df.loc[(df.even_press == '-'), 'even_press'] = 0
df.loc[(df.even_wind == '-'), 'even_wind'] = 0
df
```

```python
df['day_temp_f'] = df['day_temp'] * 9/5 + 32
df
```

```python
def filter_temp(df: pd.DataFrame, value: int) -> pd.DataFrame:
    """Creates a new DataDrame where the temperature >= value

    Args:
        df (pd.DataFrame): start DataFrame
        value (int): start temperature

    Returns:
        pd.DataFrame: filtered DataFrame
    """
    pf = df[df['day_temp'] >= value]
    return pf

filter_temp(df, df['day_temp'].mean())
```

```python
def filter_data(df: pd.DataFrame, data_start: str, data_end: str) -> pd.DataFrame:
    """Creates a new DataDrame where the data_start <= data <= data_end

    Args:
        df (pd.DataFrame): start DataFrame
        data_start (str): start data
        data_end (str): end data

    Returns:
        pd.DataFrame: filtered DataFrame
    """
    df['data'] = pd.to_datetime(df['data'], format='%Y-%m')
    pf = df.loc[(df['data'] >= data_start) & (df['data'] <= data_end)]
    return pf

filter_data(df, '2010-01-10', '2020-10-21')
```

```python
from typing import Tuple

def group_month(df: pd.DataFrame, month: int) -> Tuple[pd.DataFrame, int]:
    """Group the DataFrame by month and calculate the average value of temperature.

    Args:
        df (pd.DataFrame): Start dataframe
        month (int): Number of month

    Returns:
        Tuple[pd.DataFrame, int]: filtered DataFrame and average value of temperature
    """
    df['month'] = df['data'].dt.month
    pf = df[df['month'] == month]
    return pf, pf['day_temp'].mean(), pf['day_temp_f'].mean()

group_month(df, 8)
```

```python
def draw_weeks_stat(pd: pd.DataFrame, month: int, year: int) -> None:
    df['year'] = df['day'].dt.year
    pf = df[df['year'] == year]
    pf['month'] = df['day'].dt.month
    sf = pf[pf['year'] == year]

    list1=[]
    list2=[]

    x =  pf['exchange_rate']
    x.reset_index(drop=True, inplace=True)
    l = len(x)

    y1 = sf['exchange_rate'].mean()
    y2 = sf ['exchange_rate'].median()

    for i in range(l):
        list1.append(y1)
        list2.append(y2)     
   
    plt.ylabel('Exchange rate')
    plt.xlabel('Day')
    plt.plot(x, color='blue', linestyle='-.', linewidth=1, label='Exchange rate') 
    plt.plot(list1, color='green', linestyle='-', linewidth=1, label='Deviation from avg') 
    plt.plot(list2, color='red', linestyle='--', linewidth=1, label='Deviation from median')
    plt.legend()
    plt.title('Сurrency exchange rate changes over all time')
    plt.show()

draw_weeks_stat(df, 12, 1993)
```
